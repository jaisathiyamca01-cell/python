#### 4.3 To use try catch and handle exception  for number divisable zero

```python
try:
    num1=int(input('Enter a number 1: '))
    num2=int(input('Enter a number 2: '))
    print(num1/num2)
except ZeroDivisionError as e:
    print('error:',e)
```

#### 4.4 To get count of how many vowels in a string

```python
vowels={'a','e','i','o','u'}
vowelsInUserInput=set()
userInput=input('Enter user input:')
vowelsCount=0;
for ch in userInput:
    try:
        if vowels.issuperset(ch)==True:
            vowelsInUserInput.add(ch)
    except ValueError as e:
        print(e)

print(f'Vovels count {len(vowelsInUserInput)}')
```
