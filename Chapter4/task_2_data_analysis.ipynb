## 2. Merge functions example

```python
customers = pd.DataFrame({
    "customer_id": [101, 102, 103, 104],
    "customer_name": ["Sathiya", "Selva", "Ashok", "Inba"],
    "city": ["Chennai", "Madurai", "Coimbatore", "Madurai"]
})

orders = pd.DataFrame({
    "order_id": [1001, 1002, 1003, 1004, 1005],
    "customer_id": [101, 102, 102, 105, 103],
    "order_amount": [500.0, 750.0, 300.0, 900.0, 1200.0]
})
```

```python
pd.merge(left=customers,right=orders)
```

```python
pd.merge(left=customers,right=orders,on='customer_id',how='inner')
```

```python
pd.merge(left=customers,right=orders,on='customer_id', how='outer',indicator=True)
```

```python
pd.merge(left=customers,right=orders,on='customer_id',how='left',indicator=True)
```

```python
pd.merge(left=customers,right=orders,on='customer_id',how='right',indicator=True)
```
