```python
#guess the numbers
import random
print("Guess number between 1 to 50")
secret_number=random.randint(1,50)


for i in range(1,5):
    user_number=int(input("enter number"))
    if secret_number < user_number :
        print("The given number greater")
    elif secret_number > user_number:
        print("The given number is less")
    else :
        print(f"Found number in {i} items")
        break;

if secret_number==user_number :
    print("You found number")
else :
    print("You miss to found number")
    print(f"secret number : {secret_number}")
```
